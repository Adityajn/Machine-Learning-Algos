```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(color_codes=True)
```

```python
tips = sns.load_dataset("tips")
```

```python
#visulalize regression
#regplot accepts the x and y variables in a variety of formats including simple numpy arrays,pandas Series objects
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
"""
regplot vs lmplot
both are same but with a little difference
replot accepts data in form of numpy.array, pandas series
lmplot accepts data in form of strings
Other than this input flexibility, regplot() possesses a subset of lmplot()’s features
"""
```

```python
sns.lmplot(x="size", y="tip", data=tips);
```

```python
"""
x_jitter - only for visualization, don not chage regression line
"""
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
```

```python
"""
A second option is to collapse over the observations in each discrete bin,
plot an estimate of central tendency along with a confidence interval:
"""
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean);
```

```python
"""
Fitting different kinds of models
"""
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"), ci=True);
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),ci=True);
```

```python
"""
The above model is definately not a good model
Lets fit a polynomial model in this dataset
"""
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None);
```

```python
#When the y variable is binary, simple linear regression also “works” but provides implausible predictions
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           y_jitter=.03)
```

```python
#The solution in this case is to fit a logistic regression, such that the regression line shows the estimated probability
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03);
```

```python
#The best way to separate out a relationship
#to plot both levels on the same axes and to use color to distinguish them

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#To add another variable, you can draw multiple “facets” 
#which each level of the variable appearing in the rows or columns of the grid:
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time",data=tips);
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg");
```

```python
sns.pairplot(tips, x_vars=["total_bill", "size"], y_vars=["tip"],
             height=5, aspect=.8, kind="reg");
```
